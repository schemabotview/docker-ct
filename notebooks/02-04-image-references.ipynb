## Image references, tags & inspection

When you wrote `docker run nginx`, you used a hugely abbreviated name. The full **image reference** is:

```
[registry-host[:port]/][namespace/]repository[:tag][@digest]
```

Every piece has a default, filled in left to right:

- `nginx` → `library/nginx` (the `library/` namespace holds Docker Hub's official images).
- `library/nginx` → `docker.io/library/nginx` (Docker Hub is the default registry).
- no tag → `:latest` (the *tag named* latest — **not** "the newest").
- fully pinned: `docker.io/library/nginx:1.27.0@sha256:abc123…` — registry, namespace, repo, tag, *and* digest.

| Reference | Means |
|---|---|
| `ubuntu` | `docker.io/library/ubuntu:latest` |
| `myorg/api:v2.1.0` | `docker.io/myorg/api:v2.1.0` |
| `ghcr.io/myorg/api:v2.1.0` | same image on GitHub Container Registry |
| `nginx@sha256:abc…` | a specific immutable digest, whatever tag points there |

**Tag vs digest** is the crux: a tag like `1.27.0` *can* be re-pointed at a new image by the publisher (mutable); a digest cannot (immutable). Tags are policy, not technology — Docker doesn't care what you write after the colon, but teams converge on patterns: **semver** (`1.4.2`), **sliding semver** (`1.4`, `1`), **git SHA** (`a1b2c3d`), **date**, or **channel** (`stable`, `edge`). Rule: **never deploy `:latest` to production** — pin a version, or a digest for the strictest reproducibility.

**Opening the box — without running it:**

- **`docker image inspect <ref>`** dumps the full JSON config: layers with digests, default command, env vars, exposed ports, user, working dir, labels, architecture, OS. The source of truth for "what will this do when I run it?"
- **`docker image history <ref>`** lists the layers in order, which instruction created each, and its size — gold for understanding *why* an image is as big as it is.
