## The default `bridge` network

Install Docker and you get three networks out of the box, plus a Linux bridge `docker0` on the host:

```bash
docker network ls
# NAME      DRIVER    SCOPE
# bridge    bridge    local     ← the default; every container lands here
# host      host      local
# none      null      local
```

Every container started **without** `--network` joins the default `bridge`. It works — but it has two limitations that make it the wrong choice for anything real:

1. **No embedded DNS.** Containers on the default bridge can only reach each other by **IP**. Run `--name web` and `--name api` and, from inside `web`, the hostname `api` does *not* resolve — you'd have to dig the IP out of `docker inspect`. Brittle, since IPs change on restart.
2. **Weak isolation.** *Every* container without an explicit `--network` shares this one bridge. So an unrelated container you spun up for a quick test can talk to your dev Postgres, just because both landed on `docker0`.

The takeaway is short: **the default bridge is a historical default — don't rely on it.** The moment two containers need to talk, create a **user-defined bridge** (next section), which fixes both problems: name-based DNS and per-network isolation. You'll essentially never choose the default bridge on purpose; it's just where containers fall when you forget to say otherwise.
