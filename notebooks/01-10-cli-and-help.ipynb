## Getting help & the docker CLI shape

You won't memorise every flag of every command — you don't need to. The CLI's shape is regular, and its built-in help is excellent, so you can always find your way.

**The shape.** Every command is `docker <object> <verb> [flags] [args]`. Once you know the objects, you can guess commands: containers, images, volumes, networks, plus `system`, `builder`, `buildx`, `compose`, `context`, `secret`, `service`, `stack`, `swarm`. Verb names repeat across objects — `ls`, `inspect`, `rm`, `prune` — so `docker volume ls` and `docker network ls` behave exactly as you'd expect.

**Built-in help, three levels deep:**

- **`docker --help`** — the top-level list of objects and commands.
- **`docker <object> --help`** — every verb for that object, e.g. `docker container --help`.
- **`docker <object> <verb> --help`** — every flag of one command. `docker run --help` is the single most-read help page in Docker.

For deeper reference: **docs.docker.com** (current, searchable), and every official image on Docker Hub ships a README covering its tags, env vars, and volumes.

**Cleaning up.** A few hours in, you'll have dozens of exited containers and old images. Remove them targeted or in bulk:

```bash
docker rm <c>        docker rmi <image>       # one at a time
docker system prune          # stopped containers, dangling images, unused networks
docker system prune -a       # also ALL unused images
docker system prune -a --volumes   # also unused volumes — DATA LOSS risk
```

The habit that saves you: when stuck, run `<command> --help` first (instant — you'll find the flag), then reach for the docs. That closes module 01 — you can now run, list, inspect, and clean up containers, and you understand the client → daemon → image → container → registry map the rest of the course fills in.
