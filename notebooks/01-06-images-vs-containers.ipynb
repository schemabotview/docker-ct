## Images vs. containers

Two words you'll use every day, and they are *not* the same thing.

An **image** is a frozen, read-only template — a snapshot of a filesystem plus a little metadata (what command to run, which user, which environment variables). It doesn't *do* anything; it just sits there as a blueprint. Think of it as a class in object-oriented programming.

A **container** is a running (or stopped) instance of an image — like an object created from that class, with its own state. You can start many containers from one image, and each gets its own thin **writable layer** stacked on top of the image's read-only layers. Writes go to that layer; the image underneath never changes.

| Concept | Image | Container |
|---|---|---|
| OOP | Class | Object |
| Cooking | Recipe | The dish you cooked |
| Software | The installer | The installed, running program |

The verbs map cleanly onto the split:

- You **build** or **pull** an *image*.
- You **run** an image to create a *container*.
- You **stop**, **start**, and **rm** (remove) *containers*.

One image, many containers — and because they share the same read-only layers on disk, ten containers from one image cost almost nothing extra to store. That copy-on-write layering is what makes containers cheap; module 02 goes deep on how images and layers are built, and module 03 goes deep on running containers. For now, just keep the two words straight: **an image is the template; a container is a running instance of it.**
