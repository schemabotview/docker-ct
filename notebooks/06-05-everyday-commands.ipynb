## The everyday commands

Compose commands mirror the single-container verbs from module 03, but operate on the **whole project** at once. The ones you'll live in:

| Command | What it does |
|---|---|
| `docker compose up [-d]` | Create + start every service (`-d` = detached). |
| `docker compose up -d --build` | Rebuild `build:` images first, then start. |
| `docker compose down [-v]` | Stop + remove containers and the project network. `-v` also removes named volumes. |
| `docker compose ps` | List the project's containers and their state. |
| `docker compose logs [-f] [svc]` | Tail logs — all services, or one; `-f` to follow. |
| `docker compose exec svc sh` | Shell into a running service (Compose's `docker exec`). |
| `docker compose run --rm svc cmd` | Spin up a **new** one-off container to run `cmd` (migrations, tasks). |
| `docker compose restart / stop / start` | Cycle services without re-creating them. |
| `docker compose config` | Print the fully-resolved YAML (after variables + overrides). |

Two distinctions worth burning in early:

- **`exec` vs `run`.** `exec` runs a command **inside an already-running** service container (debug a live web app). `run` starts a **fresh, separate** container of that service for a one-shot job (`docker compose run --rm web python manage.py migrate`) — it doesn't touch the running one. Reach for `exec` to inspect, `run` to do a task.
- **`config` is your debugger.** Before puzzling over why a variable didn't substitute or an override didn't apply, run `docker compose config`: it prints the final YAML Compose actually uses, with all `.env` substitution and file merging resolved. It's the fastest way to see what Compose *thinks* you asked for.

Everything else — `build`, `pull`, `stop`, `start` — does exactly what the single-container verb did, scoped to the project.
