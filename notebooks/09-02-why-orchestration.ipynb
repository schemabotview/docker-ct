## Why orchestration & Swarm vs Kubernetes

Compose runs a stack on one host. Five questions expose its ceiling — and define what an orchestrator is for:

- What if **one host isn't enough** capacity?
- What happens when that **host dies**? (Compose has no answer — you're down until you rebuild it.)
- How do you do a **rolling deploy** without dropping traffic?
- How do you keep **secrets** off disk on every machine?
- How do services on **different hosts** talk by hostname?

An **orchestrator** answers all five. It manages a pool of hosts (a **cluster**), accepts a **desired state** ("3 replicas everywhere"), and *continuously reconciles* reality toward it: a node dies and it launches replacements elsewhere; a new image lands and replicas roll forward one at a time; a service publishes a port and traffic is routed to a healthy replica. Declarative desired state, reconciled by a control loop — that's the whole idea.

Two orchestrators matter, and both run the **same OCI images** you already build:

| | **Swarm** | **Kubernetes** |
|---|---|---|
| Install | `docker swarm init` — one command | `kubeadm` / EKS-GKE-AKS — nontrivial |
| CLI | `docker` — the client you know | `kubectl` — a separate model |
| Concepts | service, task, stack, secret | pod, Deployment, Service, Ingress, PVC… (dozens) |
| Networking | built-in overlay + routing mesh | CNI plugins — your choice |
| Ecosystem | small, Docker-shipped | vast (Helm, operators, meshes) |

**Swarm is simpler and near-instant; Kubernetes is more powerful and the industry default for non-trivial production.** This module teaches Swarm — it's DCA content, and its small model makes the *concepts* of orchestration clear before you meet Kubernetes' complexity.
