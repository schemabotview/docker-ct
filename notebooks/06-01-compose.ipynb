## Docker Compose & multi-container apps

Everything so far ran one container at a time with `docker run`. But a real app is rarely one container — it's a web server *and* a database *and* a cache *and* maybe a worker, wired together with a network, volumes, shared env vars, and a startup order. Launching that by hand is a page of `docker run` flags you retype every time. This module is the fix: **describe the whole stack once, in a file, and bring it up with one command.**

That file-and-one-command tool is **Docker Compose**, and this module lives in the **Orchestrate → Compose** corner of the map. The core idea:

```
   compose.yaml  ──►  docker compose up  ──►  web + db + cache
   (declare the stack)                        (network, volumes, order, all wired)
```

The whole module is really one YAML file learned piece by piece:

- **The shape** — `services:` (your containers) plus optional `networks:`, `volumes:`, and per-service keys (`image`/`build`, `ports`, `environment`, `depends_on`).
- **Making it robust** — real startup order via healthchecks, tiered networks, layered config for dev vs prod, and the gotchas (project names, scaling limits).

One thing to be clear about up front: **Compose is a single-host tool** — brilliant for local development and small deployments, but it is *not* a cluster orchestrator. Running containers across many machines is Swarm (module 09) or Kubernetes. Compose runs N containers, declaratively, on one Docker daemon — and that covers an enormous amount of real work.
