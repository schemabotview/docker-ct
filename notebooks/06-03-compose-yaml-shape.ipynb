## The `compose.yaml` shape

The conventional filename is **`compose.yaml`** (the older `docker-compose.yml` still works). The minimum valid file is tiny:

```yaml
services:
  web:
    image: nginx:alpine
```

One service, `web`, running `nginx:alpine`. `docker compose up` starts it, auto-creating a network for the project and joining `web` to it. Everything else is optional structure layered on top:

```yaml
name: my-app            # project name (default = directory name)

services:               # the containers — the heart of every file
  web:
    image: ...          # OR build: ...   (what to run)
    ports: ...          # host:container publishes
    environment: ...    # env vars
    volumes: ...        # mounts
    depends_on: ...     # startup order
    healthcheck: ...    # readiness probe
    restart: ...        # restart policy

networks:               # named networks (one default is auto-created)
  backend: {}

volumes:                # named volumes (like `docker volume create`)
  pgdata: {}
```

The shape to internalise: **`services:` is the only required block**, and each service is essentially a declarative `docker run` — the flags you learned in modules 03–05 (`-p` → `ports`, `-e` → `environment`, `-v` → `volumes`, `--restart` → `restart`) become YAML keys. The top-level `networks:` and `volumes:` blocks declare shared resources that services then reference by name.

A few keys that pay for themselves early: **`build:`** to build an image instead of pulling one, **`depends_on:`** for order, and **`healthcheck:`** for *real* readiness — each gets its own section. But the mental model is simply: a file of services, each a `docker run` in YAML, plus the named resources they share.
