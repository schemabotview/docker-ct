## Node management & the realities of Swarm

Running a cluster means maintaining its **nodes** without dropping traffic — and Swarm's availability states make that routine. `docker node ls` shows every node; each has an availability:

- **`active`** — accepting new tasks (default).
- **`pause`** — keeps its tasks but takes no new ones.
- **`drain`** — takes no new tasks **and** Swarm reschedules its existing tasks elsewhere.

That last one is the **rolling host upgrade** workflow — take a node down for a kernel/OS patch without losing replicas:

```bash
docker node update --availability drain nodeX   # tasks move off
# ... reboot / patch nodeX ...
docker node update --availability active nodeX   # back in rotation
```

Combined with rolling service updates, you can march a whole cluster through an OS upgrade with **zero downtime**. You also grow/shrink the control plane in place — `docker node promote workerX` (add to the manager set, e.g. 1→3 for HA) and `docker node demote managerX` — and remove a node with `docker swarm leave`.

**The honest state of Swarm in 2026.** *What works:* setup is unbeatable (`init` + a `join` per node = a cluster in two minutes), the model is small and stays close to single-host Docker, and for small-to-medium, single-region deployments it does the job at a fraction of Kubernetes' operational overhead — edge/IoT clusters often pick Swarm or k3s for exactly this. *What doesn't:* the ecosystem stagnated after Kubernetes won — new tools (operators, meshes, GitOps) target K8s first; Swarm's update model lacks K8s' richness (no canary/blue-green natively); and the hiring pool is smaller. **For the DCA exam, Swarm is core** — know the flags, the routing mesh, secrets, and placement. **For your career, knowing both is best** — Swarm makes the *concepts* of orchestration clear, which is the perfect on-ramp to Kubernetes.
