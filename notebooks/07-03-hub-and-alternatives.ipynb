## Docker Hub & the alternatives

**Docker Hub** (`docker.io`) is the default registry — `docker pull nginx` pulls from Hub without you naming it. It's free for public images but **rate-limited for anonymous pulls** (roughly 100 pulls per IP per 6 hours since 2020); authenticating or paying raises the cap. That rate limit is the single most common reason teams reach for an alternative or a cache (section 9).

The registries you'll actually meet:

| Registry | Hostname | Best for |
|---|---|---|
| **GitHub Container Registry** | `ghcr.io` | anything on GitHub; free for public, tight Actions integration — the common OSS choice |
| **AWS ECR** | `<acct>.dkr.ecr.<region>.amazonaws.com` | AWS workloads; IAM-integrated auth |
| **Google Artifact Registry** | `<region>-docker.pkg.dev` | GCP workloads (replaces `gcr.io`) |
| **Azure ACR** | `<name>.azurecr.io` | Azure workloads |
| **Quay** | `quay.io` | Red Hat / OpenShift ecosystem |
| **Harbor** | self-hosted | on-prem private registry with RBAC, scanning, replication — the standard self-hosted option |
| **Distribution (`registry:2`)** | self-hosted | the plain OSS reference server; testing or air-gapped networks |

Because they **all speak the same OCI protocol**, you can mix them freely: build in CI, push to GHCR for storage, mirror to ECR for fast prod-side pulls, and run a local cache — the same `docker push`/`pull` commands work against every one, only the hostname in the reference changes. The choice is mostly about *where your workloads run* (cloud-native auth) and *whether you need self-hosting* (Harbor / `registry:2`), not about the mechanics, which are identical.
