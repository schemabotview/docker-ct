## Config instructions ‚Äî `WORKDIR`, `ENV`, `ARG`, `EXPOSE`, `USER`

The supporting cast. Each is small; together they shape how the image behaves when run. Most are **metadata-only** ‚Äî they change the config blob, not the filesystem, so they add no layer of bytes.

**`WORKDIR /path`** ‚Äî sets the working directory for subsequent `RUN`, `CMD`, `ENTRYPOINT`, `COPY`, `ADD`. Like `cd` between instructions, but it *persists* (a `RUN cd /path` does not). Creates the dir if missing.

**`ENV KEY=value`** ‚Äî an environment variable visible at **build time and run time**, pinned into the image:

```dockerfile
ENV PYTHONUNBUFFERED=1 PIP_NO_CACHE_DIR=1
```

**`ARG NAME[=default]`** ‚Äî a **build-time-only** variable, *not* present at run time. Passed with `docker build --build-arg NAME=value`. Good for a dependency version or a build flag:

```dockerfile
ARG APP_VERSION=dev
RUN echo "building $APP_VERSION"
```

**`EXPOSE 8080`** ‚Äî *documentation* that the container listens on 8080. **It does not publish the port** ‚Äî publishing is a run-time `-p` (module 05). `docker run -P` uses it to auto-publish.

**`USER appuser`** ‚Äî the user (and optional group) that later `RUN`s and the main process run as. Default is `root`, which you **don't** want in production; pair with a `RUN useradd appuser` earlier.

Two more you'll meet: **`LABEL key=value`** adds metadata (maintainer, source repo, version) read by `docker inspect` and registries; **`HEALTHCHECK`** gives Docker a command to test liveness, surfacing a `healthy`/`unhealthy` status (Compose can gate startup on it ‚Äî module 06).
