## Listing & inspecting — `ps`, `images`

Once you've run a few containers, you need to *see* what's on your machine. Modern Docker organises commands as **object then verb**:

```
docker <object> <verb> [flags] [args]
        │        └── what to do
        └── container · image · volume · network · ...
```

So `docker container ls` lists containers and `docker image ls` lists images. For the most-used objects Docker keeps short top-level aliases — `docker ps` is `docker container ls`, `docker images` is `docker image ls`. You'll see the short forms everywhere.

**Listing containers — `docker ps`:**

```bash
docker ps        # only RUNNING containers
docker ps -a     # ALL containers, including stopped/exited
```

The columns that matter: `CONTAINER ID`, `IMAGE`, `STATUS` (e.g. `Up 3 minutes` or `Exited (0) 1 min ago`), `PORTS`, and `NAMES`. A container that vanished from `docker ps` isn't gone — it exited; `-a` reveals it.

**Listing images — `docker images`:** shows `REPOSITORY`, `TAG`, `IMAGE ID`, and `SIZE`, so you can see what's cached and what it costs on disk.

**Looking closer — `inspect`:** `docker inspect <name-or-id>` dumps the full JSON for a container or image — every mount, network, env var, and config value. Pipe it through a filter, or use `--format` to pull one field:

```bash
docker inspect --format '{{.State.Status}}' web
docker inspect --format '{{.NetworkSettings.IPAddress}}' web
```

And to see disk usage at a glance, **`docker system df`** — like Unix `df`, but for images, containers, and volumes. These read-only verbs — `ps`, `images`, `inspect`, `system df` — are how you orient before you change anything.
