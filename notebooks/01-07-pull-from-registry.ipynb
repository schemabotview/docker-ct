## Pulling from a registry

In that first `docker run`, the daemon said *"Unable to find image locally"* and went and fetched it. Where from? A **registry** — a server that stores and serves images. The default is **Docker Hub**; you can also pull from private registries or cloud ones (ECR, GCR, GHCR).

You can pull explicitly, without running anything:

```bash
docker pull nginx:1.27
docker pull ubuntu:24.04
```

The daemon contacts the registry, downloads the image's layers (skipping any you already have), and stores them in your local cache. A later `docker run nginx:1.27` then starts instantly — no network needed.

**Reading an image reference.** The full name has four parts, most of them optional:

```
        registry-1.docker.io / library / nginx : 1.27
        └── registry ──────┘  └ namespace ┘ └name┘ └tag┘
```

- **Registry** — defaults to Docker Hub if omitted.
- **Namespace** — the owner. Official images live under `library/` (which you never have to type). `myorg/app` names an organisation's image.
- **Name** — the repository.
- **Tag** — a label for a version, defaulting to **`latest`** if omitted. `latest` is just a conventional tag, **not** "the newest" magically — pin real versions in production.

So `nginx` really means `docker.io/library/nginx:latest`. Pulling is the download half of distribution; **`docker push`** is the upload half, and module 07 covers registries, tags, digests, and running your own end to end.
