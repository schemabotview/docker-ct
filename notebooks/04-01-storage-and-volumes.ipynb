## Storage, volumes & bind mounts

So far every container has been disposable — `docker rm` and it's gone, writable layer and all. That's exactly what you want for a stateless web server, and exactly what you *don't* want for a database. This module is about **making data outlive the container**: how storage attaches to a container, and which of the three mechanisms to reach for.

The whole module lives in the **Resources → Volumes** corner of the map, and turns on one move: take a path *inside* the container and **mount** something else there — something that lives outside the throwaway writable layer.

Docker gives you three flavours of mount, and choosing between them is 90% of the module:

- **Volumes** — Docker-managed storage, the **default** for persistent app data (a database's files, uploads). Independent of any container's lifecycle.
- **Bind mounts** — a specific host path mapped straight in. The **local-dev** workhorse: mount your source code so edits show up live.
- **`tmpfs`** — a RAM-backed scratch area that never touches disk. For secrets and high-throughput temp data.

```
        ┌──────── Container ────────┐
        │  /var/lib/postgresql/data │◄── volume     (persistent, Docker-managed)
        │  /app/src                 │◄── bind mount (a host path)
        │  /run/secrets             │◄── tmpfs      (RAM only)
        └───────────────────────────┘
```

By the end you'll know the default (volumes), the dev tool (binds), the escape hatch (`tmpfs`), how to spell them (`-v` vs `--mount`), and how to back them up, share them, and dodge the classic UID and Mac/Windows gotchas.
