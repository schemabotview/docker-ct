## Foreground, detached & interactive — `-d`, `-i`, `-t`

These three flags get combined constantly (`-it`, `-d`), and people memorise the combinations without knowing what each one does. Separate them and they're simple.

**`-d` — detached.** Without it, `docker run` *attaches* your terminal to the container's stdout/stderr: output streams to your terminal and `run` blocks until the container exits (Ctrl-C sends `SIGINT` to it). With `-d`, the container runs in the **background** and `run` immediately returns its ID; output is still captured for `docker logs`, just not streamed.

- **Use `-d`** for long-lived services (web servers, databases) — you want them in the background.
- **Skip `-d`** for one-shot commands (a script, a build step) — you want to watch the output and have `run` block until done.

**`-i` — interactive (keep stdin open).** Without it, the container's stdin is closed immediately, so any program that reads stdin sees EOF and exits — fine for `nginx`, broken for `bash`. With `-i`, what you type is piped to the container's stdin.

**`-t` — allocate a TTY.** Gives the container a pseudo-terminal, so you get a prompt, line editing, colours, and job control — the feel of a real shell rather than a raw pipe.

The famous combinations follow directly:

- **`-it`** — an interactive shell: `docker run -it ubuntu bash`. You need *both* — `-i` to send keystrokes, `-t` for the terminal feel.
- **`-d`** alone — a background service: `docker run -d nginx`.
- **`-dit`** — background *but* with stdin + TTY held open, so you can `docker attach` or `exec` into it later.

Rule of thumb: **`-it` when a human will interact; `-d` when nothing should be attached.**
