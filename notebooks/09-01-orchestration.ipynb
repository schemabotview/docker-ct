## Orchestration with Swarm

Compose (module 06) got you to "a multi-container app on **one** host." That's genuinely useful — and it hits a wall: one host is a single point of failure and a capacity ceiling. This module is about crossing that wall with an **orchestrator** — a system that runs your containers across a *pool* of hosts, keeps them alive when a machine dies, and rolls out new versions without downtime. The orchestrator we use is **Docker Swarm**: built into Docker, driven by the same `docker` CLI, and the one the DCA exam tests.

The module lives in the **Orchestrate → Swarm** region of the map, and turns on one idea: **declare a desired state, and let the cluster reconcile reality to it.**

- **A cluster of nodes.** Managers run the control plane (scheduling, state via raft); workers run the containers (called **tasks**).
- **Services, not containers.** You ask for "3 replicas of image X"; Swarm schedules tasks, replaces them when nodes die, and rolls updates one at a time.
- **Cluster-wide networking.** An **overlay** network lets services on different hosts talk by name, and the **routing mesh** publishes a port on *every* node.

```
   docker service create --replicas 3 web   ──►  [ Swarm: manager schedules tasks ]
                                                    task → Node A · Node B · Node C
   node dies → manager reschedules → back to 3/3
```

By the end you'll init a cluster, deploy services and stacks, do rolling updates, wire overlay networking and the routing mesh, and manage secrets, placement, and nodes — with a clear-eyed view of where Swarm fits versus Kubernetes.
