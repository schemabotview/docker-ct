## The three mount types

Docker gives you three ways to mount storage over a container path. They differ in **where the bytes live** and **whose lifecycle governs them** — get this table into your head and the rest of the module is detail.

| Type | Lives where | Lifecycle | Best for |
|---|---|---|---|
| **Volume** | Docker-managed area (`/var/lib/docker/volumes/`) | Independent of any container; deleted only by `docker volume rm` | **Persistent app data** — DB files, uploads, anything that must outlive a container |
| **Bind mount** | Any path on the host filesystem | The host path's lifecycle — Docker never touches it | **Local development** (mount your source in), config files, host log access |
| **`tmpfs`** | In host RAM | Container only — gone when it stops | **Sensitive scratch** (keep secrets off disk), high-throughput caches |

The rule that decides almost every case:

> **Use a volume by default. Use a bind mount when you need a specific host path. Use `tmpfs` when the data must never hit disk.**

Why *volume by default*? Because a volume is a **first-class Docker object** — you name it, `inspect` it, back it up, and move it, without caring where on disk it sits, and it's decoupled from any container. A bind mount ties you to an exact host path (breaks on a teammate's machine); `tmpfs` vanishes on stop. So volumes are the safe default, and you step down to the other two only for their specific jobs.

The rest of the module unpacks each type in turn — named (and anonymous) volumes, bind mounts, then `tmpfs` — followed by how to spell, share, and back them up.
