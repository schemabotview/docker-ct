## What is an image, really?

In module 01 we said an image is "a frozen, read-only template — a filesystem snapshot plus a bit of metadata." True, but let's be precise. **A Docker image is two things, tied together by a third:**

1. **An ordered stack of filesystem layers.** Each layer is a tarball of file *changes* — additions, modifications, deletions — relative to the layer below. The bottom layer might be an entire Debian root filesystem; the next "install Python"; the next "copy your app in."
2. **A JSON config blob** describing how to run it — the default command (`CMD`), entrypoint, working directory, environment variables, exposed ports, the user to run as, labels, and build history.
3. **A manifest** ties them together: it lists each layer's digest and the config's digest, and the manifest itself is digested (SHA-256). *That digest is the image's identity.*

So when you say `nginx:1.27.0` — a **tag** — you're really pointing at a digest like `sha256:abc123…`. The tag is a mutable nickname a publisher can re-point; the **digest is the immutable truth**. Deploy by digest when you need certainty.

This design — **content-addressable** layers — is what makes images cacheable. Two images that share a base layer share the *same bytes on disk*. Pull `python:3.12-slim`, then something built on it, and the overlap downloads once, not twice.

**What an image is *not*:**

- **Not a running thing.** It does nothing until you `run` it into a container. Image = class; container = object.
- **Not a VM disk.** No kernel, no boot — just a filesystem plus run metadata.
- **Not a single file.** It's a set of layers + config + manifest, usually living in a registry and your local cache, not a `.img` you copy around (though `docker save` can bundle one into a tarball).
