## Running & inspecting containers

Modules 01–02 got you from "what is a container" to "how images are built." This module is the day-to-day craft in between: **running containers deliberately, and seeing inside them when something's wrong.** You already met `docker run` in module 01 — here we take it apart flag by flag, then cover the tools that let you observe, adjust, and debug a container that's already up.

The whole module lives in one region of the map — the **Containers** box on the Docker host — and circles three questions:

- **How do I start it the way I want?** `docker run` is the busiest command in Docker, and its flags fall into clean buckets: attachment (`-d`, `-i`, `-t`), identity (`--name`, `--hostname`, `--rm`), environment (`-e`), and resource limits (`--memory`, `--cpus`). Learn the buckets and the command stops being a wall of options.
- **How do I keep it alive?** The lifecycle (created → running → stopped) plus **restart policies** that bring a service back after a crash or a host reboot.
- **How do I see what it's doing?** `logs` for its output, and `exec` / `inspect` / `stats` / `top` to look inside a running container — the debugging loop you'll run a hundred times a day.

```
   docker run ─► [ Container: running ]
                    │  logs · exec · inspect · stats   ← observe
                    │  update · cp                     ← adjust
                    ▼  stop / rm / restart             ← lifecycle
```

By the end you can launch a container exactly as intended, keep it running, and open it up when it misbehaves — without ever rebuilding the image.
