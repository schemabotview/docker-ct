## Moving files & snapshotting — `docker cp`, `docker commit`

Two commands reach into a container's filesystem from the outside — one to move files, one to freeze the whole thing into an image. Both touch the container's **mount namespace** (its private filesystem view from module 01).

**`docker cp`** copies files or directories between host and container, and works on **running *and* stopped** containers:

```bash
docker cp ./config.yaml web:/app/config.yaml     # host → container
docker cp web:/var/log/nginx/access.log ./        # container → host
docker cp web:/etc/nginx/. ./nginx-conf/          # a whole directory
```

The container path is **absolute**; the host path may be relative. This is for **ad-hoc** moves — pulling out a log, dropping in a config to test. It is **not** persistent storage: anything you `cp` in vanishes when the container is removed. Data that must survive belongs in a **volume** (module 04).

**`docker commit`** snapshots a container's current filesystem into a new image:

```bash
docker commit lab my-ubuntu-with-python
```

It works — but **you should almost never use it.** The result is *opaque*: no record of how it was built, no Dockerfile to review, no diff against the base. It's the "save game" of images — convenient, but you've lost the recipe.

- **OK for:** exploration (save a poking-around state before you destroy the container), or a forensic snapshot of a crashed container to debug later.
- **Not OK for:** anything you'll rebuild, share with a teammate, or ship to production.

For all of those, write a **Dockerfile** (module 02) — a reviewable, reproducible recipe beats an opaque snapshot every time. That principle closes the module: prefer declared, rebuildable images and observable, disposable containers over hand-tweaked, mutable ones.
