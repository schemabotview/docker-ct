## Registries, tags & distribution

You can build images (module 02) and run them (module 03), but so far they've lived only on your machine. **Distribution** is how an image gets from your laptop to a teammate, a CI runner, or a production server — and the thing in the middle is a **registry**. This module is that whole story: what a registry is, how push and pull work, how to name and pin images so you deploy exactly what you tested, and how to run and scale your own.

The module lives in the **Registry** column of the map and turns on a few ideas:

- **A registry is content-addressable storage.** It holds **blobs** (layers + config, keyed by SHA-256 digest) and **manifests** (JSON that lists them). A **tag** is just a mutable pointer to a manifest digest. Push and pull only move the blobs you don't already have.
- **Names and pins.** The full image reference — `registry/namespace/repo:tag@digest` — decides *which* registry, *which* image, and whether you get a moving tag or an immutable digest. Getting this right is how "works in staging" becomes "works in prod."
- **Running your own.** Docker Hub is the default, but you can run a private `registry:2`, put a **pull-through cache** in front to survive rate limits, and keep it healthy with garbage collection and vulnerability scanning.

```
   build ──► docker push ──► [ Registry ] ──► docker pull ──► run
             (blobs + manifest, tag → digest)      (elsewhere)
```

By the end you'll push to any registry, read an image reference precisely, pin production by digest, and stand up a cache and a private registry of your own.
