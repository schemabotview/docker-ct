## Networking & port publishing

A container in isolation isn't much use — real apps talk: a web server to its database, an API to a cache, the outside world to your service. This module is how that traffic flows. It sits in the **Resources → Networks + Ports** corner of the map and answers two questions: **how do containers reach each other**, and **how does outside traffic reach a container.**

The mental model underneath everything: a container (on a normal network) gets its own **network namespace** — its own interfaces, IP, routing table, and firewall rules — plugged into a Linux **bridge** on the host. From inside it looks like a tiny isolated machine on a network.

Two moving parts carry the module:

- **Networks — container-to-container.** Attach containers to the same **user-defined bridge** and they resolve each other by name (embedded DNS) with no port mapping at all. Docker ships several drivers (`bridge`, `host`, `none`, `overlay`, `macvlan`), but `bridge` covers ~99% of single-host work.
- **Ports — world-to-container.** A container's ports are private by default; you **publish** one with `-p host:container` to let outside traffic in — which under the hood is just an iptables rule.

```
   outside ──► host:8080 ──►[ -p ]──► container:80        (publish)
   api ───────────────► db  (by name, same bridge)        (DNS, no -p)
```

By the end you'll pick the right network, let containers find each other by name, publish ports safely, and know what `-p` really does when it misbehaves.
