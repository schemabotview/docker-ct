## Why Compose

Picture a normal web app: a web server, a Postgres database, a Redis cache, and a background worker. Launched by hand, that's four `docker run` lines — plus everything that ties them together:

- a user-defined network connecting them,
- two named volumes (Postgres data, Redis dump),
- environment variables shared by web and worker,
- a startup order (the DB must be up before web),
- port mappings (only web should be exposed),
- and the same flags repeated across every `docker run`.

You *can* wrap that in a bash script. Teams have. It's miserable to maintain — one edit means re-reasoning about four commands and their order.

**Docker Compose replaces the script with one declarative file.** A single `compose.yaml` describes the whole stack; `docker compose up` brings it all up, `docker compose down` tears it all down, and the file is checked into git so a teammate clones and runs the identical stack. You describe the *desired state* — these services, this network, these volumes — and Compose figures out the commands.

**What Compose is — and isn't.** It's a **local-development and single-host deployment** tool: it runs N containers on **one** Docker daemon, declaratively. It is **not** a cluster orchestrator — spreading containers across many machines with self-healing and rolling updates is Swarm (module 09) or Kubernetes. Keep that boundary clear and Compose is exactly the right tool for the huge middle ground: dev environments, CI, and small single-box deployments.

**One naming note:** *Compose v1* was the old standalone `docker-compose` (hyphenated) Python binary — end-of-life. *Compose v2* is the modern **`docker compose`** subcommand (a space, built into the CLI). Everything here is v2; if you see the hyphen in an old tutorial, mentally translate it.
