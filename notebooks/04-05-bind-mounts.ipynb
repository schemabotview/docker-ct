## Bind mounts

A **bind mount** maps a path on your host directly into a container — no Docker-managed storage, no copying. The container sees the host's actual bytes.

```bash
docker run -d --name web \
  -v /home/me/projects/site:/usr/share/nginx/html:ro \
  nginx:alpine
```

This serves the host directory as Nginx's web root: edit a file on the host, refresh the browser, see the change. The **`:ro`** makes it read-only inside the container — a good default when the container shouldn't write to your source tree.

**The classic dev workflow** is what bind mounts are for:

1. Keep your source on the host (where your editor and git live).
2. Run the app in a container with that source bind-mounted in.
3. Restart on change, or use a hot-reloading runtime (`nodemon`, `flask --reload`, `uvicorn --reload`).

The trade-off: a bind mount **hardcodes the host path**, so the same `docker run` won't work on a teammate's machine unless their layout matches. That's why production deploys use volumes, not binds, and Compose files (module 06) use **relative** paths that resolve against the project root.

**Two gotchas:**

- **No image-content copy.** Unlike a fresh named volume, a bind mount *silently masks* whatever the image had at that path — mount over `/app` and the image's `/app` disappears behind the host's (possibly empty) directory.
- **Ownership surprises.** Files the container writes are owned by the container's UID on the host — the classic "why is this owned by root?" that section 10 unpacks.

Rule of thumb: **bind mounts for code you're actively editing; volumes for data you're keeping.**
