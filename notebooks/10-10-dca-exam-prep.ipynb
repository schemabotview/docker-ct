## The DCA exam, prep tactics & where to go next

The whole course has been quietly building to this: the **Docker Certified Associate** — and its six domains map almost one-to-one onto the modules you've done.

**The exam** (administered by Mirantis): **55 questions**, **90 minutes**, online-proctored, ~65–70% to pass, valid 2 years. The domains and their weights are just this course, reorganised:

| Domain | Weight | Modules |
|---|---|---|
| **Orchestration** | ~25% | 09 (Swarm, services, stacks, updates) |
| **Image creation & registry** | ~20% | 02, 07 |
| **Installation & configuration** | ~15% | 10 (daemon.json, drivers) |
| **Networking** | ~15% | 05 |
| **Security** | ~15% | 08 |
| **Storage & volumes** | ~10% | 04 |

So preparing is really *consolidating the course you've done* — orchestration and images together are ~45% of the exam.

**Memorise** (there are no man pages in the exam): the shape of `docker run` and its common flags; volume vs bind vs tmpfs semantics; when containers resolve each other by name and the Swarm routing-mesh model; the Dockerfile quirks (`CMD` vs `ENTRYPOINT`, `COPY` vs `ADD`, `ENV` vs `ARG`); services vs tasks and the rolling-update flags; the Compose/stack `deploy:` keys. **Look up** the rarely-used `daemon.json` keys and profile paths. And above all **practise, don't just read** — spin up a real 3-node Swarm (Multipass or Lima VMs) and run `service create / scale / update / rollback` and `stack deploy` until they're muscle memory.

**Where to go next.** The natural path is **Kubernetes** — it orchestrates the *same* OCI images with a richer model, and module 09's mapping makes the jump concrete (Service → Deployment, Task → Pod, Stack → Helm chart): **CKAD** (app-level) is the right cert after DCA, then **CKA** (cluster), then **CKS** (security). Or go **cloud-specific** (ECS/EKS, Cloud Run/GKE, AKS) if your work lives on one platform. Either way, you now hold the whole model — image → container → network → volume → registry → hardened → orchestrated — and can install, configure, debug, and certify it. That's the entire Docker platform, and you've built it up one map at a time.
