## Images, layers & the Dockerfile

Module 01 gave you the whole map — client, daemon, image, container, registry. This module zooms into one corner of it: **where images come from**. Every container you ran started from an image someone built; now you learn to build your own, and to understand what's actually inside one.

Three ideas carry the whole module, and they're worth seeing together before we go deep on each:

- **An image is a stack of layers.** Not one blob — an ordered set of read-only filesystem layers, each recording "what changed relative to the layer below." A union filesystem merges them into the single `/` a container sees. Because layers are content-addressed by digest, images that share a base **share the same bytes on disk** — that's what makes them cacheable and cheap to ship.
- **A Dockerfile is the recipe that builds those layers.** A plain text file of instructions — `FROM` a base, `RUN` commands, `COPY` your code in, then metadata like `CMD`. `docker build` executes it top to bottom, and *most instructions add a layer.* Learn to read a Dockerfile and you can predict exactly what an image contains and why it's the size it is.
- **The build cache rewards good ordering.** Because each instruction's layer is cached and reused until something it depends on changes, *the order of your instructions decides your build speed.* Put the rarely-changing steps (install dependencies) before the often-changing ones (copy source), and rebuilds go from minutes to milliseconds.

```
  Dockerfile ──build──►  Image (layer stack)  ──run──►  Container
   FROM · RUN            L1 · L2 · L3 · config           writable layer
   COPY · CMD            ↑ cached & shared by digest      on top
```

By the end you'll build a real Flask image, understand every line of it, and know how to make it small and fast with multi-stage builds and BuildKit.
