## AppArmor & SELinux

**AppArmor** and **SELinux** are the two **Mandatory Access Control (MAC)** systems Linux distros ship. They sit *on top of* ordinary file permissions and capabilities, deciding which paths, capabilities, and operations a process may perform — a policy the process (even as root) cannot override. Where capabilities and seccomp constrain *privileges* and *syscalls*, MAC constrains *access to resources* by policy.

**AppArmor** (Ubuntu, Debian) is **path-based**. Docker ships a **`docker-default`** profile applied automatically when AppArmor is enabled — it stops containers writing to most of `/proc`, mounting filesystems, and raw networking. Override with a custom profile when you must:

```bash
docker run --security-opt apparmor=my-profile myorg/api
```

**SELinux** (RHEL, Fedora, CentOS) is **label-based** — *type enforcement* tags every file and process, and only matching labels may interact. Docker sets container labels via `--security-opt label=...`.

The practical guidance here is different from the earlier sections: **you usually don't touch either.** They're applied by default, and the defaults are correct. You reach for a custom profile only when an audit asks "why does this container have access it doesn't need?" — then you tighten. Two things worth internalising: which system you get is **distro-dependent** (AppArmor on Debian-family, SELinux on Red Hat-family), and both are **defense in depth** — an extra wall that catches what a dropped capability or seccomp rule missed. If you have no specific reason to override, leave the defaults on and move to the higher-leverage moves: read-only rootfs and blocking privilege escalation, next.
