## Security, secrets & hardening

Module 07 ended on scanning — *detecting* what's wrong in an image. This module is the other half: **shrinking the attack surface** so there's less to go wrong, and handling **secrets** so credentials don't leak. It's the difference between a container that merely runs and one you'd put on the internet.

The module lives across the **Security** block and the **Container**'s kernel primitives, and it's organised as a threat model followed by a stack of defences:

- **Know the boundary.** A container is a *process-isolation* mechanism, **not** a hard security boundary — it shares the host kernel. Every defence here is about making a breakout harder and less useful, not about pretending the wall is unbreakable.
- **Reduce privilege, layer by layer.** Don't run as **root**; remap or go **rootless**; drop **capabilities**; filter syscalls with **seccomp**; add MAC (**AppArmor**/**SELinux**); make the **rootfs read-only** and block privilege escalation. Each is a small, independent step, and together they're the "hardened baseline."
- **Handle secrets properly.** Env vars are **not** secret — use mounted **secret files**. And prove provenance with **signing**, **SBOMs**, and attestations.

```
   [ Container ]  ── drop caps · seccomp · non-root · read-only ──►  smaller surface
        │
     secrets (files, not env)  +  signed images / SBOM   ──►  trusted supply chain
```

By the end you'll build the hardened service: pinned by digest, non-root, no capabilities, read-only, secrets as files — the config you can defend in a review.
