## Publishing ports

A container's processes bind ports **inside their own namespace**, and the bridge is private — so by default nothing outside the host can reach them. To let traffic in, you **publish** a port with `-p host:container`:

```bash
docker run -d -p 8080:80 nginx:alpine
```

That reads "take whatever arrives on the **host's** port 8080 and forward it to the **container's** port 80." Now `curl http://localhost:8080` hits the Nginx inside. The direction is the classic mix-up: **host port on the left, container port on the right.**

The variations you'll actually use:

- **`-p 8080:80`** — bind on **all** host interfaces (`0.0.0.0`) — reachable from the whole network.
- **`-p 127.0.0.1:8080:80`** — bind **only on localhost**. Reachable from your laptop, not the wider network. **Use this for any dev service that needn't be remote** — much safer than exposing it to the LAN.
- **`-p 8080:80/udp`** — choose the protocol (default is TCP).
- **`-p 80`** — host port omitted → Docker picks a **random free** host port; read it back with `docker port <container>` or the `PORTS` column of `docker ps`.

**Publish vs. `EXPOSE`.** Recall `EXPOSE` from module 02 is just *documentation* in the image — it publishes nothing. Two related shortcuts tie them together: **`-P`** (capital) auto-publishes **every `EXPOSE`d port** to a random host port (handy for a quick test, rarely for real use).

The safety habit worth forming now: **default to `127.0.0.1:`-bound publishes** for anything local, and only bind `0.0.0.0` when a service genuinely needs to be reachable from other machines. An accidental `-p 5432:5432` on a database is how dev Postgres ends up open to the internet.
