## Install, configure, troubleshoot & DCA prep

You've come the whole way — from "what is a container" to a hardened, orchestrated cluster. This final module is the operator's half of the story: **standing Docker up properly, configuring the daemon, and debugging it when things break** — plus how to turn all of it into a **DCA** certification. Where the earlier modules built *up* the platform, this one steps *back down* to the engine at the center of the map: `dockerd → containerd → runc`, the daemon stack everything else rides on.

The module runs in three movements:

- **Install & configure.** The right way to install Docker Engine (official repos, not the distro's stale package), the post-install steps that make it pleasant (the `docker` group, enable-on-boot), and the one file that tunes the daemon — **`/etc/docker/daemon.json`** — including storage, logging, and runtime drivers.
- **Troubleshoot.** A repeatable method for the three things that break: the **daemon** won't start, a **container** or **build** misbehaves, or the **network** won't pass packets. Each has a small, memorable set of first commands.
- **Certify.** What the **DCA** exam actually tests, what to memorise vs. look up, and where to go next (Kubernetes, cloud certs).

```
   install → daemon.json → [ dockerd → containerd → runc ] → troubleshoot → DCA
```

The final section is a synthesis: the six DCA domains map one-to-one onto modules 01–09, so preparing for the exam is really just consolidating the course you've already done. By the end you can install, configure, debug, and certify — a working Docker operator.
