## The five built-in drivers

Docker ships several network drivers; five are worth knowing, and one of them does almost all the work.

| Driver | What it does | When to use |
|---|---|---|
| **`bridge`** | The default. Each container gets a private IP on a host bridge; outbound is NAT'd, inbound only via `-p`. | **~99% of single-host work.** |
| **`host`** | Container shares the **host's** network namespace — no isolation, binds host interfaces directly. | Max throughput; raw sockets / host `iptables`. |
| **`none`** | Namespace with only `lo` — no connectivity at all. | Sandboxed jobs; forensic analysis. |
| **`overlay`** | Spans **multiple hosts** over VXLAN so containers on different machines share a virtual network. | Swarm clusters (module 09). |
| **`macvlan` / `ipvlan`** | Container gets its own MAC/IP on the **physical LAN** — looks like a real machine, no NAT. | Legacy LAN apps, IoT bridges — niche. |

The shape to remember: **`bridge` is the workhorse** — private IP, NAT out, publish to come in — and it's what you should assume unless a specific need pushes you off it. `host` and `none` are the two extremes (all of the host's network, or none at all), and `overlay` and `macvlan` are for multi-host and physical-LAN cases you'll meet later.
