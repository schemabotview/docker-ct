## Naming, hostname & `--rm`

Two flags control how a container is *identified*, and one controls whether it *sticks around* after it exits.

**`--name NAME`** gives the container a stable name you'll use to refer to it. Skip it and Docker invents a two-word name like `quirky_borg` or `vibrant_einstein`. Auto-names are fine for throwaways; name anything you'll talk to more than once:

```bash
docker run -d --name web nginx:alpine
docker stop web        # much nicer than `docker stop 4f8a9c...`
```

Names are **unique per host** — you can't have two containers named `web`, even if one is stopped. You'd `docker rm web` first, or pick another name. (This is a common "name already in use" error after a crash: the old stopped container still holds the name.)

**`--hostname NAME`** sets the hostname *inside* the container — what `hostname` returns from within, and what the app may put in its logs or service registration. It defaults to the container's short ID. Note this is separate from the container's `--name`, which is a daemon-side label, not something the process sees.

**`--rm`** auto-removes the container when it exits. Without it, every `docker run` leaves an `Exited` container behind forever — until you `docker rm` it or `docker system prune` (the graveyard you clean up in module 01). With it, the container is garbage-collected on exit.

- **Use `--rm`** for one-shot, interactive commands: `docker run --rm -it ubuntu bash`. No clutter in `docker ps -a`.
- **Skip `--rm`** for services you may want to `start` again, or whose exited state you want to inspect after a crash.

One catch: **`--rm` and `--restart` are mutually exclusive** — a container can't both auto-remove and auto-restart. Pick one.
