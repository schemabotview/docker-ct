## Installing Docker & the daemon

You only learn Docker by running containers, so first get a working Docker on your machine. Whatever the route, the goal is the same: a running **daemon** (`dockerd`) on a Docker host, and a `docker` client that can talk to it. There are four common paths, roughly by ease.

**Option 1 — Docker Desktop (easiest on a laptop).** Free for personal use, paid for large companies. macOS, Windows, Linux. Install, launch, and `docker` is on your `PATH`. It bundles Compose and a Kubernetes toggle. The catch: on macOS/Windows it quietly runs a small **Linux VM** — because containers need a *Linux* kernel — so host↔container file sync can be slow.

**Option 2 — Docker Engine on Linux (the real thing).** On a Linux box or cloud VM, install the engine directly — native, no VM, fastest. The abbreviated install (module 10 has the hardened version):

```bash
curl -fsSL https://get.docker.com | sh
sudo usermod -aG docker $USER    # run docker without sudo
newgrp docker
```

**Option 3 — Colima (open-source Desktop alternative).** macOS/Linux. Spins up a small Lima VM and runs the engine inside it. Free, lightweight, CLI-only: `brew install colima docker && colima start`.

**Option 4 — a cloud VM.** A small Linux droplet/EC2 instance with the engine installed, used over SSH — the most production-like, at a few dollars a month.

**Verify the install.** Two commands tell you it's healthy:

- **`docker version`** — prints *both* the client and the server (daemon) versions. Seeing both means the daemon is running and you can reach it.
- **`docker info`** — the daemon's configuration: storage driver, logging driver, kernel version, container counts, default registry.

If `docker version` shows a client but errors on the server, the daemon isn't running (or you lack permission to its socket) — that's the single most common first-day snag.
