## Your first container — `docker run`

The Docker equivalent of "hello world":

```bash
docker run hello-world
```

You'll see output like:

```
Unable to find image 'hello-world:latest' locally
latest: Pulling from library/hello-world
Status: Downloaded newer image for hello-world:latest

Hello from Docker!
This message shows that your installation appears to be working correctly.
```

That output is dense — and *every* container you ever run follows this same sequence, so read it slowly:

1. **`Unable to find image 'hello-world:latest' locally`** — the daemon checked your local image cache for `hello-world` tagged `latest`. Not there.
2. **`Pulling from library/hello-world`** — so the daemon fetched it from **Docker Hub**, the default registry. `library/` is the namespace for official images.
3. **`Status: Downloaded newer image ...`** — the image now lives in your local cache; next run skips the download.
4. **`Hello from Docker!`** — the daemon created a **container** from that image, ran its one command (which prints the message), and the container exited.

So a single `docker run` did four things: **resolve** the image name → **pull** it if missing → **create** a container → **start** it. That is the whole platform in one line: client → daemon → registry → image → container.

`docker run` is really `create` + `start` fused together. A few flags you'll reach for immediately:

- **`-d`** — detached: run in the background and return your prompt.
- **`-it`** — interactive with a TTY: for a shell, e.g. `docker run -it ubuntu bash`.
- **`--name web`** — give the container a stable name instead of a random one.
- **`--rm`** — auto-remove the container when it exits, so it doesn't linger as stopped.
