## What Docker is — and the shape of the whole platform

Ask ten people what Docker is and you'll get ten answers, all partly right. Let's untangle it — and in the process lay out the map this whole course zooms into.

**Start with the thing Docker exists to run: the container.** The cleanest one-line definition: *a container is an isolated process on a host, running its own filesystem and seeing its own network, while sharing the host's kernel.* When you start one, the Linux kernel gives it its own **namespaces** (a private view of processes, mounts, and network) and its own **cgroups** (limits on CPU, memory, and I/O).

Notice what a container is **not**: it is not a small virtual machine. There is no guest OS inside it and no second kernel to boot — every syscall goes straight to the host kernel. That's why a container starts in **milliseconds** and weighs **megabytes**, where a VM takes seconds to boot and weighs gigabytes.

**Docker is the friendly front end on top of those kernel primitives.** But "Docker" isn't one program — it's a platform with a few moving parts, and knowing how they fit together is the whole game:

- **The client** — the `docker` command you type. It sends API requests and does no real work itself.
- **The daemon (`dockerd`)** on the **Docker host** — the long-running engine that receives those requests and does the work: building images, running containers, managing storage and networks. Under it sits a stack — `dockerd` hands off to `containerd`, which uses `runc` to actually spawn the container with the kernel.
- **Images and containers** — an **image** is a frozen, read-only template (a filesystem snapshot plus a little metadata); a **container** is a running instance of one. You *build* or *pull* an image, then *run* it to create a container.
- **The registry** — where images live so you can share them: **Docker Hub**, or a private registry. You *push* images up and *pull* them down.

The mental picture — and it's the map the rest of this course zooms into:

```
  docker CLI  ──►  Docker Host                       ◄──►  Registry
  (client)         dockerd → containerd → runc              Docker Hub
                   builds images · runs containers          private registry
                        │
                 Author (Dockerfile) · Resources (volumes / networks / ports)
                 Orchestrate (Compose · Swarm)
```

**Why it's worth your time:** every cloud runs containers (ECS, Cloud Run, AKS); Kubernetes orchestrates them; CI/CD pipelines build inside them; `docker compose up` replaced "install five things on your laptop." Learning Docker is the floor under all of it — and a prerequisite to Kubernetes. This course takes you from zero to the **DCA** (Docker Certified Associate) exam by module 10.
