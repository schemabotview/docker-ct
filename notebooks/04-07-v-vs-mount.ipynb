## `-v` short form vs `--mount` long form

Docker has two ways to spell a mount, and they mostly do the same thing — but they fail differently, which is the whole reason to care.

**`-v` (short, terse):** `-v <source>:<destination>[:<options>]`

The source is a **volume name**, an **absolute host path** (→ bind mount), or **omitted** (→ anonymous volume). Options are comma-separated (`ro`, `rw`, `z`/`Z` for SELinux relabel, `cached`…):

```bash
-v pgdata:/var/lib/postgresql/data      # named volume
-v /host/src:/app/src:ro                # bind mount, read-only
```

**`--mount` (long, explicit):** `--mount type=<volume|bind|tmpfs>,source=...,target=...,readonly`

```bash
--mount type=volume,source=pgdata,target=/var/lib/postgresql/data
--mount type=bind,source=/host/src,target=/app/src,readonly
--mount type=tmpfs,target=/scratch,tmpfs-size=64m
```

The trade-off is **safety vs. brevity**:

- **`-v` is shorter but ambiguous.** `-v foo:/data` is a *volume named foo*; `-v /foo:/data` is a *bind mount of /foo*. One leading slash flips the meaning — a typo silently sends data to the wrong place, and worse, `-v ./missing:/data` **auto-creates** the host dir rather than erroring.
- **`--mount` is verbose but unambiguous** — you write `type=volume` or `type=bind` explicitly, and a bad source **errors** instead of guessing. The docs recommend `--mount` for new code.

You'll meet both in Compose (module 06): short `volumes:` strings are `-v` style, long `volumes:` mappings are `--mount` style. Read either; prefer `--mount` when writing something you'll keep.
