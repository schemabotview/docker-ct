## The anatomy of `docker run`

`docker run` is the busiest command in Docker, and its full shape is:

```
docker run [OPTIONS] IMAGE[:TAG|@DIGEST] [COMMAND] [ARG...]
```

The image is required; everything in `[OPTIONS]` tunes *how* it runs, and everything after the image *overrides the default command* baked into it (the `CMD`/`ENTRYPOINT` from module 02). So `docker run ubuntu echo hi` runs `echo hi` instead of ubuntu's default.

The options look overwhelming as a flat list, but they fall into a handful of **mental buckets** — and once you see the buckets, you can guess where a flag lives:

| Bucket | Flags | Covered |
|---|---|---|
| **Attachment / lifecycle** | `-d` `-i` `-t` `--rm` `--name` `--hostname` | this module |
| **Environment** | `-e` `--env-file` `-w` (workdir) `-u` (user) | this module |
| **Resource limits** | `--memory` `--cpus` `--pids-limit` `--ulimit` | this module |
| **Restart & init** | `--restart` `--init` | this module |
| **Networking** | `-p` `-P` `--network` `--dns` | module 05 |
| **Storage** | `-v` `--mount` `--tmpfs` `--read-only` | module 04 |
| **Security** | `--cap-add` `--cap-drop` `--security-opt` `--user` | module 08 |

That's the map for the whole module: we take the attachment, environment, resource, and restart buckets here; networking, storage, and security each get their own module later. A useful reflex whenever a `docker run` line looks dense: **sort its flags into these buckets** and it reads as four or five small decisions, not one big one.
