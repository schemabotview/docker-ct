## Environment variables — `-e` and `--env-file`

Containerised apps read their configuration from **environment variables** — the 12-factor pattern, and why the same image can be dev in one run and prod in another. Docker sets them two ways.

**Inline with `-e`** — one variable per flag:

```bash
docker run -e DATABASE_URL=postgres://... -e LOG_LEVEL=info myorg/api
```

**Bulk with `--env-file`** — a file of `KEY=value` lines (`#` for comments):

```
# api.env
DATABASE_URL=postgres://user:pw@db:5432/app
LOG_LEVEL=info
```

```bash
docker run --env-file=api.env myorg/api
```

**Precedence**, lowest to highest, when the same key appears more than once:

1. `ENV` baked into the image (module 02).
2. `--env-file` files (later files beat earlier).
3. `-e` flags (rightmost wins).

A handy shorthand: **`-e VAR` with no value inherits it from your host shell** — good for passing a token without echoing it on the command line:

```bash
export GITHUB_TOKEN=ghp_...
docker run -e GITHUB_TOKEN myorg/ci-tool
```

**A caution on secrets.** Env vars are widely used for config, but they're a poor home for *secrets*: they leak through `docker inspect`, the process's `/proc` environment, and crash dumps, and child processes inherit them. Fine for a log level or a feature flag; for real credentials, use proper secrets — Compose secrets, Swarm secrets, or BuildKit secret mounts — which module 08 covers.
